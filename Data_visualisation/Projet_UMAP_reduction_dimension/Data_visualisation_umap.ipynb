# R√©duction de dimension avec UMAP

## squash_stored

```python
X = squash_stored.drop(51, axis=1)
Y = squash_stored[51]
reducer = umap.UMAP(n_neighbors=5, min_dist=0.3, n_components=3)
embedding = reducer.fit_transform(X)
```

```python
# Cr√©ation d'un DataFrame pour la visualisation
df = pd.DataFrame({'UMAP 1': embedding[:, 0], 'UMAP 2': embedding[:, 1], 'number': Y})

# Visualisation des donn√©es r√©duites avec UMAP
plt.figure(figsize=(10, 8))
sns.scatterplot(x='UMAP 1', y='UMAP 2', hue='number', data=df, palette='viridis')
plt.title('R√©duction de dimension UMAP pour le jeu de donn√©es squash_stored')
plt.show()
```
