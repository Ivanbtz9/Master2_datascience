# Algorithme de la puissance ittéré

```python
def puissance_itt(A,step=10,eps = 1e-4):
    u = np.random.randn(len(A)).reshape(-1,1)
    for k in range(step):
        u = A@u
        if np.linalg.norm(u)< eps:
            print(k)
            break
        u /= np.linalg.norm(u)
        
    return u, (A@u/u).mean()


puissance_itt(np.array([[10,8],[8,10]]) )
```

# Contrainte sur la norme 1

```python
def proxL1(vec,la):
    return np.sign(vec) * np.maximum(0,np.abs(vec)-la)
```

```python
v = np.array([3,-4,6])
proxL1(v,2)
```

```python
def projL1(vec,rds):
    if np.linalg.norm(vec,1)<=rds:
        return vec,0
    u = np.sort(np.abs(vec))[::-1]
    n = len(u)
    ukmaok = (np.cumsum(u)-rds)/np.arange(1,n+1)
    K = np.argmax(ukmaok[ukmaok<u])
    lam = ukmaok[K]
    res = proxL1(vec,lam)
    return res,lam
    
```

```python
projL1(np.array([2,5,8,0.3]),3)
```

# Algo puissance itérée modifiée

```python
def puissance_itt_modif(A,res=2,step=10,eps = 1e-4):
    u0 = np.random.randn(len(A)).reshape(-1,1)
    
    for k in range(step):
        u0 = A@u0
        if np.linalg.norm(u0)< eps:
            print(k)
            break
        u0 = projL1(u0.flatten(),res)[0].reshape(-1,1)
        u0 /= np.linalg.norm(u0,2)
        
    return u0, u0.T@A@u0


puissance_itt_modif(np.array([[10,8],[8,10]]),step=10,res=2.3)
```

```python
import numpy.linalg as npl
```

```python
data = {}
for stimulus in y.values:
    data[stimulus]= df[df['Stimulus.Name']==stimulus][df.columns[2:]]
    
```

```python
BCG = data['BCG']
BCG = (BCG - BCG.mean(axis=0))/BCG.std() 
```

```python
CorrBCG = BCG.corr()
```

```python
preciBCG = npl.inv(CorrBCG)
sns.heatmap(preciBCG)

```

```python
for i in range(len(preciBCG)):
    preciBCG[i,i] = 0
```

```python
plt.hist(preciBCG.flatten()[preciBCG.flatten()>5], bins=50, edgecolor='black')  # Augmenter le nombre de bins à 50

plt.title('Distribution des Valeurs de la Matrice de Précision')
plt.xlabel('Valeur')
plt.ylabel('Fréquence')
```

```python
for i in range(len(preciBCG)):
    for j in range(len(preciBCG)):
        if np.abs(preciBCG[i,j]) <30:
            preciBCG[i,j] = 0 
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Supposons que preciBCG est votre matrice de précision calculée précédemment
# et que BCG.columns contient les noms des colonnes correspondantes

# Création du graphe à partir de la matrice de précision
G = nx.from_numpy_array(preciBCG)

# Trouver tous les nœuds avec un degré de 0 (isolés) et les supprimer
nodes_to_remove = [node for node, degree in dict(G.degree()).items() if degree == 0]
G.remove_nodes_from(nodes_to_remove)

# Mise à jour des labels des nœuds après suppression
labels = {i: var for i, var in enumerate(BCG.columns) if i not in nodes_to_remove}

# Ajustement de la distance de répulsion entre les nœuds pour améliorer la dispersion
# Augmentez la valeur de 'k' pour éloigner les nœuds les uns des autres
pos = nx.spring_layout(G, k=4)  # Vous pouvez ajuster 'k' selon les besoins

# Dessin du graphe mis à jour avec un espacement accru
plt.figure(figsize=(12, 12))
nx.draw_networkx(G, pos, labels=labels, with_labels=True, node_color='skyblue', node_size=700, edge_color='k', linewidths=1, font_size=10)

plt.title("Graphe basé sur la Matrice de Précision (sans nœuds isolés)")
plt.axis('off')  # Cacher les axes pour une meilleure visualisation
plt.show()
```
