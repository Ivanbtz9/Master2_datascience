```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Simuler des étiquettes vraies
np.random.seed(0)  # Pour la reproductibilité
y_true = np.random.binomial(1, 0.5, 1000)

# Générer des scores de prédiction
scores_performant = np.random.normal(loc=0.8, scale=0.1, size=1000) * y_true + np.random.normal(loc=0.2, scale=0.1, size=1000) * (1 - y_true)
scores_peu_performant = np.random.normal(loc=0.6, scale=0.4, size=1000) * y_true + np.random.normal(loc=0.4, scale=0.4, size=1000) * (1 - y_true)
```

```python
# Modèle Performant
precision_performant, recall_performant, _ = precision_recall_curve(y_true, scores_performant)
display_performant = PrecisionRecallDisplay(precision=precision_performant, recall=recall_performant)
display_performant.plot(label='Modèle Performant')

# Modèle Peu Performant
precision_peu_performant, recall_peu_performant, _ = precision_recall_curve(y_true, scores_peu_performant)
display_peu_performant = PrecisionRecallDisplay(precision=precision_peu_performant, recall=recall_peu_performant)
display_peu_performant.plot(label='Modèle Peu Performant', color='red')

plt.title("Comparaison des Courbes Précision-Rappel")
plt.legend()
plt.show()
```
