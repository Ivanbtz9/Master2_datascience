## Questions 4 :

```python
def replace_random_value(df):
    for col in df.columns:
        values_col = df[col].dropna().values
        if df[col].isnull().sum() > 0: #vérifie si il ya un Nan dans la colonne
            # met une valeur aléatoire si elle est nulle 
            df[col] = df[col].apply(lambda x: np.random.choice(values_col) if pd.isnull(x) else x)
    return df
        
data1 = replace_random_value(data1)

data1.sample(4)
```

## Question 5 : 

```python
def normalisation_df(df):
    return (df -df.mean(axis=0))/df.std(axis=0)

data1 = normalisation_df(data1)

#vérification
normalisation_df(data1).mean(axis=0).round(decimals=3).abs().sample(2)
normalisation_df(data1).std(axis=0).round(decimals=3).abs().sample(2)
```

## Question 6 :

```python
 def scatter_plot_cause_effet(data,target,display=True):
    for index, row in target.iterrows(): # permet d'ittérer sur les lignes et index (data.iterrows())
        source_var = row.iloc[0]
        target_var = row.iloc[1]

        # Sélection des données pour les variables source et cible
        x = data[source_var]
        y = data[target_var]
        
        if display:
            # Création du nuage de points
            plt.figure(figsize=(8, 6))
            plt.scatter(x, y)
            plt.title(f'Nuage de points : {source_var} → {target_var}')
            plt.xlabel(source_var)
            plt.ylabel(target_var)
            plt.show()

scatter_plot_cause_effet(data1,target1)
```

## Question 8

Dans le cas où le graphe admetrait au moins un cyle, ma matrice d'adjacence ne serait pas nilpotente et donc pas semblable à une matrice triangulaire supérieure strictement. Il suffit de trouver les valeurs propres de la matrice d'adjacente pour vérifier qu'elle est bien nilpotente. Nous pouvons aussi vérifier dans un premier temps que la matrice d'adjacence est triangulaire stricte. 

```python
def find_cycle(target,data):
    unique_values = pd.unique(target.values.ravel('K'))
    #Le paramètre 'K' dans la méthode .ravel() spécifie que l'aplatissement doit être fait en suivant l'ordre de stockage en mémoire
    mapping = {value: i for i, value in enumerate(unique_values)}
    for col in target.columns:
        target[col] = target[col].map(mapping)
        
    N = len(data.columns)
    ajacent_matrix = np.zeros((N,N))
    for index,row in target.iterrows():
        try:
            i = row.iloc[0]
            j = row.iloc[1]
            ajacent_matrix[i,j]=1
        except:
            print(row.iloc[0][1:],row.iloc[1][1:])
    # Calculer les valeurs propres de la matrice d'adjacence avec leur multiplicité
    eigenvals = np.linalg.eigvals(ajacent_matrix)
    # Vérifier si l'une des valeurs propres est nulle (ou très proche de zéro) et renvoie un Booleen 
    all_eigen_values_null = np.isclose(eigenvals, 0).all()
    
    if all_eigen_values_null:
        print(f"Il n'y a pas de cycle dans le graphe")
    else:
        print("Il y a au moins un cycle dans le graphe")
        
    return eigenvals,all_eigen_values_null
        
find_cycle(target1,data1)
```

```python
def pipeline(data,target,display=True):
    data = replace_random_value(data)
    data = normalisation_df(data)
    scatter_plot_cause_effet(data,target,display=display)
    find_cycle(target,data)

for i in range(1,6):
    data = pd.read_csv(os.path.join(data_path,f'data{i}.csv'))
    target = pd.read_csv(os.path.join(data_path,f'target{i}.csv'))
    print(f'data{i}')
    pipeline(data,target,display=False)
```
