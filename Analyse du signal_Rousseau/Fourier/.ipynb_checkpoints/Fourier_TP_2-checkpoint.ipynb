```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import time    
```

```python
# Import datasets
link="C:/Users/msi/Desktop/M2_Data Science/Traitement du signal/KinectDataGreenhouse"

Box1_Temoin =pd.read_csv(link+"/Box1_Temoin.csv")
Box2_Stress_hydrique=pd.read_csv(link+"/Box2_Stress_hydrique.csv")
Box3_stress_salin=pd.read_csv(link+"/Box3_stress_salin.csv")
Box4_Temoin=pd.read_csv(link+"/Box4_Temoin.csv")
```

```python
# plot of datasets
plt.plot(Box1_Temoin.X, Box1_Temoin.Y, label='line 1')
plt.plot(Box2_Stress_hydrique.X, Box2_Stress_hydrique.Y)
plt.plot(Box3_stress_salin.X, Box3_stress_salin.Y)
plt.plot(Box4_Temoin.X, Box4_Temoin.Y)
plt.legend()
plt.show()
```

```python
def devibyday(df):

    j=0
    Y_pred=[]
    for i in range(len(df)//96):
       
       modele = LinearRegression()
       X=np.array(Box1_Temoin["X"][j:j+96])
       Y=np.array(Box1_Temoin["Y"][j:j+96])
       modele.fit(X.reshape(-1, 1), Y)
       y_pred=list(modele.predict(X.reshape(-1, 1)))
       Y_pred=Y_pred+y_pred
       j=j+96
    df=df.head((len(df)//96)*96)
    df["Y_pred"]=Y_pred
    df["Y_1"]=df["Y"]-df["Y_pred"]
    return df

devibyday(Box1_Temoin)
       

```

```python
plt.plot(devibyday(Box1_Temoin).X,devibyday(Box1_Temoin).Y_1)
```

```python
plt.plot(devibyday(Box1_Temoin).X,devibyday(Box1_Temoin).Y)
plt.plot(devibyday(Box1_Temoin).X,devibyday(Box1_Temoin).Y_pred)
```
