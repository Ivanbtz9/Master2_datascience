```python
import os,sys
import time, datetime
import re
import pandas as pd 
import numpy as np

path = 'log'

pattern = r"PID: (\d+) - Execution : (.*?) is (\d+\.\d+)"

df = pd.DataFrame(columns=["PID","Jobs", "Time_ex"])

with open(path,'r') as f:
    for l in f:
        match = re.search(pattern, l)
        if match:
            df.loc[len(df)] = [match.group(1), match.group(2), match.group(3)]
df
        
```

```python

df['Time_ex'] = pd.to_numeric(df['Time_ex'], errors='coerce')

df_sort = df.sort_values('Time_ex')

df_sort['Time_ex'] = df_sort['Time_ex'].map(lambda x: np.floor(x/min(df_sort['Time_ex'])))  

df_sort
```

```python
gb_PID = df.groupby(['PID'])
gb_PID['Time_ex'].mean() 
```
