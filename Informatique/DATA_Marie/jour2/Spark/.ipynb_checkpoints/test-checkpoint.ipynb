```python
from pyspark import SparkContext
sc = SparkContext("local","First App")
```

```python
lines = sc.textFile('/home/ibotcazou/Bureau/Master_data_science/Master2_datascience/Informatique_charbonel/DATA_Marie/Jour1/text2.txt')
word_count = lines.flatMap(lambda line: line.split())
print(word_count)
word_count2 = word_count.map(lambda word: ( word,1))
print(word_count2)
word_count3 = word_count2.reduceByKey(lambda a,b: a+b)
print(word_count3)
res = word_count3.collect()

for word,cpt in res:
    print(word,cpt)
```
