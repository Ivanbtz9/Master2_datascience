# Cours

*La fonction map*

```python
carre = lambda t:t*t

n = 15
x = map(carre,range(n)) 
#est un ittérateur qui prend moins de place qu'une liste
# il ne peut être lu qu'une seule fois

print('premier passage')
for i in x:
    print(i)
print('second passage')
for i in x:
    print(i)

```

```python
dir(x)
```

**filter**

```python
f1 = lambda t:t%2==0 #version Fct = lambda x:F(x)

x = map(carre,range(n))
y = filter(f1,x)
#il ne reste plus ien dans x car il a servi 
#y est un ittérateur
for i in y:
    print(i)
```

**Reduce**

```python
from functools import reduce

f2 = lambda t,u: t+u

x = map(carre,range(n))
y = filter(f1,x)
z = reduce(f2,y)
print(z)
```

# Exercice 2

```python
import random
import numpy as np

l = [3,5,7,9,6,2]

def calcul_moy_var(l):
    s1 = lambda x,y: (x+y)
    
    f_map = lambda t: 1
    mapper1 = map(f_map,l)
    n = reduce(s1,mapper1)
    
    m = reduce(s1,l)/n

    carre = lambda x : (x-m)**2

    cm = map(carre,l)
    z =  reduce(s1,cm)/n
    
    return m,z

print(calcul_moy_var(l))


```

```python
from random import uniform
from statistics import mean, variance

n = 1000
X = [uniform(2.5, 10.0) for i in range(n)]
print("Résultats attendus")
print(f"Moyenne : {mean(X):.3f}, Variance : {variance(X)*(n-1)/(n):.3f}")

print("\nRésultats obtenus")
res = calcul_moy_var(X)
print(f"Moyenne : {res[0]:.3f} - Variance : {res[1]:.3f}")
```
