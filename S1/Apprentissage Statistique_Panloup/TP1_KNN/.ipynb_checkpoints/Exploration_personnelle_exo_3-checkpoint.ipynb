```python
n=1200
r = 1000

sample = df_train.sample(n)

Dn = sample.sample(n-r) #choisir n lignes pour lancer

indice_supprimer = Dn.index

Dr = sample.drop(index = indice_supprimer, axis=0)
```

```python

#Groupement par numéro
group_number = Dn.groupby(0)

group_number.sum()
```

```python

density = {}

for i in range(10):
    #Faire une moyenne sur les données et normaliser pour avoir une densité empirique
    density[f'{i}'] = group_number.mean().iloc[i].values/group_number.mean().iloc[i].sum()

nb = list(density.keys())

for i in range(10):
    plt.plot(density[f'{i}'],label=nb[i],c=f"#{i}{9-i}99")
    plt.title(f"densité empirique du nombre {i}")
    plt.legend()
    plt.show()
```

# Test avec des distances sur des densités

```python
#https://fr.wikipedia.org/wiki/Distance_statistique

def variation_totale(t):
    return np.abs(t+1)/2

def Kullback_Leibler(t):
    eps = 10**(-8)
    return t*np.abs(np.log(t+eps))

#Familles de divergences qui me donne une métrique entre deux densités
def D_f(u,v,f):
    eps = 10**(-7)
    return (f(u/(v+eps))*v).sum()
    
#Nous allons choisir la fonction de Kullback_Leibler car donne une bonne séparation entre les nombres
print("D_f(density['0'],density['0'],Kullback_Leibler) = ",D_f(density['0'],density['0'],Kullback_Leibler))
print("D_f(density['0'],density['3'],Kullback_Leibler) = ",D_f(density['0'],density['3'],Kullback_Leibler))
```

# Autres métriques

## Kullback_Leibler

```python
y = Dr[0]
y_pred = np.zeros(len(y))
X = Dr.drop(columns=0,axis=1)

X_density = np.array(X)/np.array(X.sum(axis=1)).reshape((len(y),1))

for i in range(len(y)):
    liste_distance = []
    
    for j in range(10):
        liste_distance.append(D_f(density[f'{j}'],X_density[i],Kullback_Leibler)) 
    a = np.argmin(np.array(liste_distance)) 
    y_pred[i] = a

    
score = (y_pred == y).sum()/ len(y)

print(f"Avec la fonction Kullback_Leibler méthode le score est de {score}") 

```

```python
y = Dr[0]
y_pred = np.zeros(len(y))
X = Dr.drop(columns=0,axis=1)

X_density = np.array(X)/np.array(X.sum(axis=1)).reshape((len(y),1))

for i in range(len(y)):
    liste_distance = []
    
    for j in range(10):
        liste_distance.append(D_f(density[f'{j}'],X_density[i],variation_totale)) 
    a = np.argmin(np.array(liste_distance)) 
    y_pred[i] = a

    
score = (y_pred == y).sum()/ len(y)

print(f"Avec la fonction variation_totale méthode le score est de {score}") 

```
